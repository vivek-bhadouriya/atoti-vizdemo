```python
import pandas as pd
from pandas.io.json import json_normalize
import datetime
import csv

# pip install atoti[jupyterlab]
# Start jupyterlab session
import atoti as tt
```

```python
# create atoti session
session = tt.create_session()
#open a sesion for browser
session.url
```

```python
# Read the csv file, it can also read from parquet data file (read_parquet)
session.load_all_data()
match_df=session.read_csv("covid_19_data.csv",keys=["SNo"])
match_df.head()
```

```python
#Create cube and define Dimension and Measures
cube=session.create_cube(match_df)
h,l,m=cube.hierarchies,cube.levels,cube.measures
cube
```

```python
# We can also add new measure in to the cube
m["Recovered#"]=tt.agg.max(match_df["Recovered"])
cube.query(m["Recovered#"],include_totals=True,levels=l["SNo"])
```

```python
m["Deaths#"]=tt.agg.max(match_df["Deaths"])
cube.query(m["Deaths#"],include_totals=True,levels=l["SNo"])
```

```python
m["Confirmed#"]=tt.agg.max(match_df["Confirmed"])
cube.query(m["Confirmed#"],include_totals=True,levels=l["SNo"])
```
